```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("OLA.csv")
```

```python
# 1. Booking Trends Over Time
plt.figure(figsize=(12, 6))
sns.countplot(x='month', data=df, order=df['month'].value_counts().index)
plt.title('Monthly Booking Trends')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# 2. Booking Distribution by Gender
plt.figure(figsize=(8, 8))
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Booking Distribution by Gender')
plt.show()
```

```python
# 3. Popular Times of the Day
plt.figure(figsize=(12, 6))
sns.countplot(x='time_of_day', data=df, order=df['time_of_day'].value_counts().index)
plt.title('Popular Times of the Day for Bookings')
plt.xlabel('Time of Day')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# 4. Distance vs. Time Taken
plt.figure(figsize=(10, 6))
sns.scatterplot(x='distance_travelled', y='time_taken', data=df)
plt.title('Distance vs. Time Taken')
plt.xlabel('Distance Travelled')
plt.ylabel('Time Taken')
plt.show()
```

```python
# 5. Reasons for Travel
plt.figure(figsize=(10, 6))
sns.countplot(x='reason', data=df, order=df['reason'].value_counts().index)
plt.title('Reasons for Travel')
plt.xlabel('Reason')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# 6. Cab Category Preferences
plt.figure(figsize=(8, 6))
sns.countplot(x='category', data=df, order=df['category'].value_counts().index)
plt.title('Cab Category Preferences')
plt.xlabel('Cab Category')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# 7. Commission and Driver Fare Analysis
plt.figure(figsize=(12, 6))
sns.boxplot(x='commission_base_cost', y='driver_base_cost', data=df)
plt.title('Commission and Driver Fare Analysis')
plt.xlabel('Commission Base Cost')
plt.ylabel('Driver Base Cost')
plt.show()
```

```python
# 8. Tax Analysis
df_tax = df[['total_tax', 'commission_base_cost', 'driver_base_cost']]
df_tax.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Tax Analysis')
plt.xlabel('Booking ID')
plt.ylabel('Amount')
plt.show()
```

```python
# 9. Total Trip Cost Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['total_trip_cost'], bins=20, kde=True)
plt.title('Total Trip Cost Distribution')
plt.xlabel('Total Trip Cost')
plt.ylabel('Frequency')
plt.show()
```

```python
# 10. Customer Ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='ratings', data=df, order=sorted(df['ratings'].unique()))
plt.title('Customer Ratings Distribution')
plt.xlabel('Ratings')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# 11. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
customer_retention = df.groupby('month')['booking id'].nunique()
customer_retention.plot(marker='o')
plt.title('Customer Retention Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Unique Customers')
plt.show()
```
